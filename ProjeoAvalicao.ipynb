```python
"""
At the command line, only need to run once to install the package via pip:

$ pip install google-generativeai
"""

from pathlib import Path
import hashlib
import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

# Set up the model
generation_config = {
 "temperature": 1,
 "top_p": 0.95,
 "top_k": 0,
}

safety_settings = [
 {
  "category": "HARM_CATEGORY_HARASSMENT",
  "threshold": "BLOCK_MEDIUM_AND_ABOVE"
 },
 {
  "category": "HARM_CATEGORY_HATE_SPEECH",
  "threshold": "BLOCK_MEDIUM_AND_ABOVE"
 },
 {
  "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
  "threshold": "BLOCK_MEDIUM_AND_ABOVE"
 },
 {
  "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
  "threshold": "BLOCK_MEDIUM_AND_ABOVE"
 },
]

model = genai.GenerativeModel(model_name="gemini-1.0-pro",
generation_config=generation_config,
safety_settings=safety_settings)

def extract_pdf_pages(pathname: str) -> list[str]:
 parts = [f"--- START OF PDF ${pathname} ---"]
 # Add logic to read the PDF and return a list of pages here.
 pages = []
 for index, page in enumerate(pages):
   parts.append(f"--- PAGE {index} ---")
   parts.append(page)
 return parts

convo = model.start_chat(history=[
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document0.pdf")
 },
 {
  "role": "user",
  "parts": ["verifique se a data no escrito deste documento √© 10/05/2024"]
 },
 {
  "role": "model",
  "parts": ["O texto fornecido n√£o cont√©m a data 10/05/2024. A √∫nica data presente no texto √© 30/03/2024."]
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document1.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document2.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document3.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document4.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document5.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document6.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document7.pdf")
 },
 {
  "role": "user",
  "parts": ["fa√ßa o mesmo"]
 },
 {
  "role": "model",
  "parts": ["Dos documentos fornecidos, os docs 4, 5 e 6 cont√™m a data 10/05/2024."]
 },
 {
  "role": "user",
  "parts": ["agora comece de novo, na proxima resposta voc√™ vai me pedir os pdf's para analisar e eu vou de passar uma data de vencimento, voc√™ vai analisar todos os pdf's e me retornar apenas os que n√£o cont√©m a data que eu disser."]
 },
 {
  "role": "model",
  "parts": ["Ok, estou pronto. Por favor, forne√ßa os PDFs para an√°lise e me diga qual data de vencimento voc√™ quer que eu exclua dos resultados."]
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document8.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document9.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document10.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document11.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document12.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document13.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document14.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document15.pdf")
 },
 {
  "role": "user",
  "parts": ["10/05/2024"]
 },
 {
  "role": "model",
  "parts": ["Os seguintes PDFs **N√ÉO** cont√™m a data 10/05/2024:\n\n* doc 8.pdf\n* doc 7.pdf\n* doc 3.pdf\n* doc 2.pdf\n* doc 1.pdf"]
 },
 {
  "role": "user",
  "parts": ["agora voc√™ vai me pedir um lote de pdf's, eu vou te enviar para analisar e voc√™ vai pedir quais grupo de argumentos que caso encontrado algo diferente ser√° apontado o nome do doc por voc√™, note que vou colocar o grupo de argumentos separado por \";\""]
 },
 {
  "role": "model",
  "parts": ["Ok, estou pronto para receber o lote de PDFs. Por favor, envie os arquivos e me diga quais grupos de argumentos voc√™ quer que eu analise. Separe cada grupo com \";\". ¬†\n\nPor exemplo: \"Data de vencimento: 10/05/2024; Valor: R$ 100,00\". \n\nAssim que eu receber os PDFs e os argumentos, vou analis√°-los e apontar o nome de qualquer documento que apresente diverg√™ncias em rela√ß√£o aos argumentos fornecidos."]
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document16.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document17.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document18.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document19.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document20.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document21.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document22.pdf")
 },
 {
  "role": "user",
  "parts": extract_pdf_pages("<path>/document23.pdf")
 },
 {
  "role": "user",
  "parts": ["10/05/2024;venci;"]
 },
 {
  "role": "model",
  "parts": ["Entendi! Voc√™ me enviou 8 PDFs e quer que eu verifique se algum deles cont√©m informa√ß√µes diferentes de \"10/05/2024\" para \"Data de vencimento\" e \"venci\" para algum outro argumento n√£o especificado. \n\nFavor me enviar os nomes dos outros argumentos que voc√™ deseja que eu analise em cada PDF, al√©m de \"Data de vencimento\". ¬†\n\nPor exemplo: \"Nome do cliente; Valor; N√∫mero da fatura\". \n\nAguardo sua resposta para prosseguir com a an√°lise."]
 },
 {
  "role": "user",
  "parts": ["10/05/2024;venci"]
 },
 {
  "role": "model",
  "parts": ["Ent√£o voc√™ quer que eu verifique se em algum dos 8 PDFs:\n\n1. **\"Data de vencimento\" √© diferente de \"10/05/2024\"**, ou\n2. **Aparece a palavra \"venci\" em qualquer lugar do documento**.\n\nCorreto?\n\nSe for isso, posso prosseguir com a an√°lise e te mostrar os resultados. üòä"]
 },
 {
  "role": "user",
  "parts": ["sim"]
 },
 {
  "role": "model",
  "parts": ["Ap√≥s analisar os PDFs, os seguintes documentos apresentam diferen√ßas em rela√ß√£o aos argumentos fornecidos:\n\n* **doc 8.pdf:** Data de vencimento 30/03/2024\n* **doc 7.pdf:** Data de vencimento 09/05/2024\n* **doc 3.pdf:** Data de vencimento 09/05/2024\n* **doc 2.pdf:** Data de vencimento 31/03/2024\n* **doc 1.pdf:** Data de vencimento 30/03/2024 \n\nNenhum dos documentos cont√©m a palavra \"venci\"."]
 },
])

!pip install pymupdf
!pip install langchain
!pip install tabulate

import tabulate


%cd /content/PDFS
from google.colab import files
arquivo = files.upload()

# Obtenha uma lista de todos os arquivos PDF na pasta
arquivos_pdf = [arquivo for arquivo in os.listdir() if arquivo.endswith(".pdf")]

import os
from langchain.document_loaders import PyMuPDFLoader

# ... (c√≥digo inicial)

# Initialize an empty dictionary to store results per PDF
results_per_pdf = {}

# Prompt the user for the arguments
arguments = input("Por favor insira os argumentos separados por virgulas: ").split(",")

# Specify the folder containing the PDFs
folder_path = "/content/PDFS"

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
  # Check if the file is a PDF
  if filename.endswith(".pdf"):
    # Load the PDF using PyMuPDFLoader
    loader = PyMuPDFLoader(os.path.join(folder_path, filename))
    pages = loader.load()

    # Initialize an empty dictionary to store results for this PDF
    pdf_results = {}

    # Check if all arguments are present in the PDF
    for argument in arguments:
        if argument in " ".join(page.page_content for page in pages):
            pdf_results[argument] = True
        else:
            pdf_results[argument] = False

    # Add the results for this PDF to the main dictionary
    results_per_pdf[filename] = pdf_results

# Generate the final report
print("**Relat√≥rio de resultados:**")

for filename, pdf_results in results_per_pdf.items():
    print(f"\n**PDF: {filename}**")

    # Check if all arguments were found
    if all(result for result in pdf_results.values()):
        print("Todos os argumentos foram encontrados.")
    else:
        print("Argumentos n√£o encontrados:")
        for argument, result in pdf_results.items():
            if not result:
                print(f"- {argument}")

# Check for arguments found in different PDFs
for argument in arguments:
    found_in_different_pdfs = False
    for pdf1, pdf1_results in results_per_pdf.items():
        for pdf2, pdf2_results in results_per_pdf.items():
            if pdf1 != pdf2 and pdf1_results[argument] and not pdf2_results[argument]:
                found_in_different_pdfs = True
                print(f"Argumento '{argument}' encontrado em '{pdf1}' mas n√£o em '{pdf2}'.")

    if not found_in_different_pdfs:
     print(f"N√£o foram encontrados argumentos em diferentes PDFs para '{argument}'.")

# ... (c√≥digo anterior para gerar os resultados)

# Crie a lista de dados
data = []
for filename, pdf_results in results_per_pdf.items():
  for argument, result in pdf_results.items():
    data.append([filename, argument, "Encontrado" if result else "Nao encontrado"])

# Converta a lista em tabela
table_str = tabulate.tabulate(data, headers=["PDF", "Argumentos", "Resultado"], tablefmt="fancy_grid")

# Imprima a tabela
print(table_str)

# ... (c√≥digo anterior para gerar a tabela)

import csv

# Abra um arquivo para escrita no formato CSV
with open("resultados.csv", "w", newline="") as arquivo_csv:
  escritor_csv = csv.writer(arquivo_csv)

  # Escreva o cabe√ßalho da tabela
  escritor_csv.writerow(["PDF", "Argumentos", "Resultado"])

  # Escreva cada linha de dados
  for linha in data:
    escritor_csv.writerow(linha)

# Imprima uma mensagem informando o usu√°rio sobre o arquivo CSV salvo
print("O arquivo CSV 'resultados.csv' foi salvo no seu Google Drive.")



```
